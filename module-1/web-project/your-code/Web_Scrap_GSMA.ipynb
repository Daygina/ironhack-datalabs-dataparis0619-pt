```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.gsmarena.com/'
```

```python
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
```

```python
site = soup.find_all('div',{'class':'news-item'})
site
```

```python
site1 = soup.find_all('h3')
site1
```

```python
titles=[title.text for title in site1]

titles
```

```python
keyword='Huawei'
string='\n'.join(string for string in titles if keyword in string)
print(string)
```

```python
substring = "Huawei"
count = string.count(substring)
print("The count is:", count)
# How many mention of Huawei brand on the website of GSMA
```
