Display the trending Python repositories in GitHub
The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

```python
url = 'https://github.com/trending/python?since=daily'
html =requests.get(url).content
soup = BeautifulSoup(html, "lxml")
```

```python
#<h1 class="h3 lh-condensed">
#<a href="/1995parham/github-do-not-ban-us">      <span class=...</path></svgthub-do-not-ban-us

trendyrepo_list = soup.find_all('h1',{'class':'h3 lh-condensed'})
trendyrepo_list
```

```python
repo_names=[]
for contenu in trendyrepo_list:  
    name = contenu.find('span').text # find tag a
    repo_names.append((name))
repo_names 
#trending python repos names
```
