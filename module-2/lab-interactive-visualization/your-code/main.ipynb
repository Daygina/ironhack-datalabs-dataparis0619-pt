## 1. Create an interactive bar chart showing total quantity and revenue by country (excluding United Kingdom) for the month of April 2011.

```python
data1=data[(data['Country'] != 'United Kingdom') & (data['InvoiceDate'] >='2011-04-01') & (data['InvoiceDate'] <='2011-04-30')]
```

```python
data1=data1.groupby('Country', as_index=False)['Quantity', 'Revenue'].agg('sum')
#data1=data1.groupby('Country', as_index=False).agg({'Quantity': 'sum', 'Revenue': 'sum'})
```

## 2. Create an interactive line chart showing quantity and revenue sold to France between January 1st and May 31st 2011.

```python
data2=data[(data['Country'] == 'France') & (data['InvoiceDate'] >='2011-01-01') & (data['InvoiceDate'] <='2011-05-31')]
```

```python
data2=data2.groupby(['InvoiceDate'], as_index=False).agg({'Quantity': 'sum', 'Revenue': 'sum'})
```

```python
data2.iplot(kind='line', x='InvoiceDate', xTitle='Dates', yTitle='Sales', title='France Sales over 5 months')
```
