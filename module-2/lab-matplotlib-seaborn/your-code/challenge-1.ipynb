#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python

fig, [ax1, ax2] = plt.subplots(1,2, figsize=(15,5))
ax1.plot(x,x**2, x, np.exp(x) )
plt.grid(True)

ax2.plot(x,x**2, x, np.exp(x)) #, 'g-', 'k^',
plt.grid(True)

ax1.set_xscale('linear')
ax2.set_yscale('log')
```

```python
fig, [ax1, ax3] = plt.subplots(1,2, figsize=(15,5))
ax1.plot(x,x**2, x, np.exp(x))
ax3.plot(x,x**2, x, np.exp(x))

ax1.set_xscale('linear')
ax3.set_yscale('symlog')
plt.tight_layout()
plt.yscale('symlog', linthreshy=0.015)
```

#### In the vehicles data set that you have downloaded, conduct exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. Please provide at least one inssight from the plot that you have drawn

```python
data = pd.read_csv('vehicles.csv') 
# examples from the guide
sns.scatterplot(x="CO2 Emission Grams/Mile", y="Highway MPG", hue="Drivetrain", data=data)
```

Scatter Plot

```python
plt.scatter(x=data["Cylinders"], y=data["Engine Displacement"])
plt.xlabel('Cylinders X axis')
plt.ylabel('Engine Displacement Y axis')
```

```python
plt.scatter(x=data["CO2 Emission Grams/Mile"], y=data["Combined MPG"])
plt.xlabel('CO2 Emission Grams/Mile')
plt.ylabel('Combined MPG')
```

Box Whisker Plot

```python
plt.boxplot(x=data["CO2 Emission Grams/Mile"], widths=0.6, patch_artist=True)
plt.xlabel('CO2 Emission Grams/Mile')
#to explore  args: positions=(10, 20) vert 
```

```python
plt.boxplot(x=data["Combined MPG"], notch=True, widths=0.9, vert=False, patch_artist=False)
plt.xlabel('Combined MPG')
#to explore  args: positions=(10, 20) vert 
```

Histogram

```python
data['Cylinders'].plot.hist(bins=20, color='teal')
```
